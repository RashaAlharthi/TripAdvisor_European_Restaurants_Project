```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load and Read the dataset
european_restaurants = pd.read_csv("cleaning_data.csv",encoding='latin1')
```

### 1st PLOT

```python
# What is the average rating if resturant gluten free or not

ax = sns.catplot(x = "gluten_free", y = "avg_rating", kind = "bar", data = european_restaurants, aspect= 1.5, palette = 'crest')
plt.subplots_adjust(top = 0.9)
ax.fig.suptitle('Avg Rating if resturant gluten free or not', fontsize = 16)
ax.set_axis_labels("Gluten Free", "Average Rating")
```

- ### Insight:


#### In this plot we plotting the relationship between gluten-free and the average restaurant rating, we find that:

If the restaurant provides gluten-free meals, the highest rating is approximately 4.5, while if it does not provide gluten-free meals, the highest rating is 4. This indicates that the restaurant’s provision of gluten-free dishes increases the positive evaluation of the restaurant.

### 3rd PLOT

```python
# which city has the most Resturant in europ

x = european_restaurants['city'].value_counts().nlargest(20)#count value
plt.figure(figsize=(16,4))#size
ax = sns.barplot(x.index, x.values, alpha=0.8,palette='crest', edgecolor='green')

plt.title("Which city has the most resturant?", fontsize = 16)
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
plt.ylabel('# number of resturant', fontsize=12)
plt.xlabel('city', fontsize=12)

#adding the text labels
rects = ax.patches
labels = x.values
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')

plt.show()
```

- ### Insight:


#### This plot is to find the Top 20 city has a restaurants in europ, we find that:

- Rome has highest number of restaurants which is 16861 resturant.
- Finale Ligure has the lowest number of restaurants which is 2685 resturant.
- All of Genoa, Budapest and Menorca have a close number of restaurants.

### 5th PLOT

```python
# What is the relationship between language and average rating?

plt.figure(figsize=(12,10))#size
ax = sns.countplot(hue='avg_rating', x='default_language', data = european_restaurants, palette='crest')#color
ax.set_ylabel('Count')
ax.set_xlabel('language')
ax.set_title('language of resturant and its Average Rating', fontsize = 16)#title

for p in ax.patches:
        width, height = p.get_width(), p.get_height()
        x, y = p.get_xy() 
        ax.text(x+width-.05, 
                y+height,
                '{:.0f}'.format(height),
                weight='bold',
                                horizontalalignment='center',
                size='small') 
                
plt.show()
```

- ### Insight:

In this plot we have compiled restaurants that are with English language and restaurants that are with all languages and count the number of each rating value, we find that:


- The resturants with English language have higher rating than othe language.
- For English language resturants 4.5  and 4 are  higher rating and the lowest 1.
- For non English language resturants 4  and 4.5 are higher rating and the lowest 1.5.

### 8th PLOT

```python
# Compute the correlation matrix
corr = european_restaurants.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(14, 10))

# Generate a custom diverging colormap
#cmap = sns.diverging_palette(230, 20, as_cmap=True)
cmap =sns.color_palette("crest", as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0, annot=True, fmt=".2f",
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

- ### Insight:


Here we have the correlation plot to describe all relations between each coloumn in our dataset, we find that:
- There is a some relation between vegetarian_friendly and vegan_options.
- There is a some relation between vegan_options and gluten_free.
- There is a strong relation between open_days_per_week and the working_shifts_per_week.
- There is a strong relation between total_reviews_count and reviews_count_in_default_language, also some some relation with average, poor and terrible.
- There is a strong relation between avg_rating with all of value, service and food ratings.
- There is a strong relation between reviews count in default language and both of very good and excellent ratings.
- There is a strong relation between poor rating and terrible rating.
- There is a strong relation between food rating and service rating.
- There is a strong relation between average of reviwes and poor rating.

### 14th PLOT

```python
#what is the Top 5 restaurants with a avg rating of 5?

# To take all resturants have average rating of 5
avg_grp = european_restaurants.groupby('avg_rating')
five = avg_grp.get_group(5)

# calculate how many times each restaurant name is appeared with an average rating of 5
top = five['restaurant_name'].value_counts().nlargest(5)

colors = sns.color_palette('crest')#color
plt.pie(top.values, labels=top.index, colors=colors, autopct='%.0f%%',  textprops={'fontsize': 16})
plt.title('Top 5 restaurants that take an average rating 5 out of 5', fontsize = 20, y=1.1)#TITLE
fig = plt.gcf()
fig.set_size_inches(5,5) 
plt.show()
```

- ### Insight:

The objective of this graph is to find the Top 5 restaurants that take an average rating 5 out of 5 multiple times, In the beginning, we take only the restaurants that got an average rating of 5, then we calculate how many times each restaurant name was repeated with this evaluation, and we took the top 5 names. We found that Boulangerie Marie Blachere is the most resturant rated by 5 in our data which indicates that this restaurant is the most quality restaurant (37% represents the frequency of rating this restaurant with an average rating of 5). Also subway, Domino's Pizza, La Dolce Vita and Cream & cioccolato are in the top five restaurants with an average rating of 5.

### 15th PLOT

```python
# To take all resturants have food rating of 5
fo_grp = european_restaurants.groupby('food')
food = fo_grp.get_group(5)

# To take all resturants have average service of 5
sr_grp = european_restaurants.groupby('service')
service = sr_grp.get_group(5)

# To take all resturants have value rating of 5
val_grp = european_restaurants.groupby('value')
value = val_grp.get_group(5)

# To take all resturants have atmosphere rating of 5
at_grp = european_restaurants.groupby('atmosphere')
atmosphere = at_grp.get_group(5)
```

```python
#what is the Top 5 restaurants with a avg rating of 5?

fig, axs = plt.subplots(2, 2)
fig.suptitle("Top 5 restaurants that take an ratings 5 out of 5", fontsize=16)

top_5_food = food['restaurant_name'].value_counts().nlargest(5)
colors = sns.color_palette('crest')#color
axs[0, 0].pie(top_5_food.values, labels=top_5_food.index, colors=colors, autopct='%1.0f%%')
axs[0, 0].title.set_text('Top 5 restaurants with a food rating of 5')#TITLE

top_5_service = service['restaurant_name'].value_counts().nlargest(5)
colors = sns.color_palette('crest')#color
axs[0, 1].pie(top_5_service.values, labels =top_5_service.index, colors=colors, autopct='%1.0f%%')
axs[0, 1].title.set_text('Top 5 restaurants with a service rating of 5')#TITLE

top_5_value = value['restaurant_name'].value_counts().nlargest(5)
colors = sns.color_palette('crest')#color
axs[1, 0].pie(top_5_value.values, labels=top_5_value.index, colors=colors, autopct='%1.0f%%')
axs[1, 0].title.set_text('Top 5 restaurants with a value rating of 5')#TITLE

top_5_atmosphere = atmosphere['restaurant_name'].value_counts().nlargest(5)
colors = sns.color_palette('crest')#color
axs[1, 1].pie(top_5_atmosphere.values, labels=top_5_atmosphere.index, colors=colors, autopct='%1.0f%%')
axs[1, 1].title.set_text('Top 5 restaurants with a atmosphere rating of 5')#TITLE
```

- ### Insight:

In the footsteps of the previous plot, we want to find the names of the restaurants that got a rating of 5 out of 5 multiple  times in each of food, service, value and atmosphere. (The percentage represents the frequency of a restaurant name with a rating of 5) we find that La Dolce Vita is the most resturant rated by 5 in each of food, service and value, It is also the second most frequent resturant in atmosphere. We note that both  La Dolce Vita and The Red Lion restaurant are in the top 5 in all ratings which indicates that they are among the restaurants that have a high level of quality.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
res = pd.read_csv("cleaning_data 2.csv",encoding='latin1')

```

PLOT 1

```python
ax = sns.catplot(x="gluten_free", y="avg_rating",kind="bar",data=res, aspect= 1.5 ,palette='crest')
plt.subplots_adjust(top=0.9)
ax.fig.suptitle('Avg Rating if resturant  gluten free or not ')
```

if resturant provide gluten free meals the highest rating is almost 4.5 while if not provide gluten free meals the highest rating 4

PLOT2

```python

x=res['city'].value_counts()#count value
x=x.sort_values(ascending=False)#sort
x=x.iloc[1:20]#top 20
plt.figure(figsize=(16,4))#size
ax = sns.barplot(x.index, x.values, alpha=0.8,palette='crest',  edgecolor='green')
plt.title("Which city has the most resturant?")
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
plt.ylabel('# number of resturant', fontsize=12)
plt.xlabel('city', fontsize=12)

#adding the text labels
rects = ax.patches
labels = x.values
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')

plt.show()
```

From sample of top 20 city we noted
Magherafelt has highest number of resturant which is 14313
And the lowest city is Finale Ligure with 2685 resturant

PLOT4

```python
plt.figure(figsize=(8,8))#size
ax = sns.countplot(hue='avg_rating', x='default_language'  , data = res, palette='crest')#color
ax.set_ylabel('Count')
ax.set_xlabel('language')
ax.set_title('language of  resturant and its is  avg Rating')#title

for p in ax.patches:
        width, height = p.get_width(), p.get_height()
        x, y = p.get_xy() 
        ax.text(x+width-.05, 
                y+height,
                '{:.0f}'.format(height),
                weight='bold',
                                horizontalalignment='center',
                size='small') 
                
plt.show()
```

the resturants with English language have higher rating than othe language 
for English language resturants 4.5  and 4 are  higher rating and the lowest 1
for non English language resturants 4  and 4.5 are higher rating and the lowest 1.5

PLOT 7

```python
# Compute the correlation matrix
corr = res.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(14, 10))

# Generate a custom diverging colormap
#cmap = sns.diverging_palette(230, 20, as_cmap=True)
cmap =sns.color_palette("crest", as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0, annot=True, fmt=".2f",
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

describe all relation between each coloumn in our dataset
